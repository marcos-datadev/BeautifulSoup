```python
import pandas as pd
```

```python
df = pd.read_csv("quotes.csv")
df
```

```python
tags = []

for palavra in df['Tags']:
    for item in str(palavra).split():
        item = item.replace(",", "")
        tags.append(item)

tags
```

```python
tags_separadas = {}
for palavra in tags:
    if palavra not in tags_separadas:
        tags_separadas[palavra] = 1
    else:
        tags_separadas[palavra] += 1

tags_separadas = dict(sorted(tags_separadas.items(), key = lambda item : item[-1], reverse = True))
tags_separadas
```
